```python
import pandas as pd
import numpy as np
from sodapy import Socrata
from datetime import datetime
pd.set_option('display.max_columns', None)

data_url='data.cityofnewyork.us'    # The Host Name for the API endpoint (the https:// part will be added automatically)
data_set='erm2-nwe9'    # The data set at the API endpoint (311 data in this case)
app_token='uumeNlIxcHODw36DzIS3qrvxu'   # The app token created in the prior steps
client = Socrata(data_url,app_token) 
```

```python
c = 1
while c<10:
    print(c," try")
    try:
        client.timeout = 100
        query = client.get(data_set, where="lower(complaint_type) LIKE '%noise%' AND date_extract_y(created_date)=2010", limit = 10000000)
        raw_311 = pd.DataFrame.from_records(query)
        print(raw_311.shape)
        display(raw_311.head())
        break
    except:
        c+=1
```

```python
for i in range(2013,2024):
    c = 1
    while c<3:
        print(i,  "year ", c," try")
        try:
            string_query = "lower(complaint_type) LIKE '%noise%' AND date_extract_y(created_date)=" + str(i)
            client.timeout = 200
            query = client.get(data_set, where=string_query, limit = 30000000)
            raw_311_new = pd.DataFrame.from_records(query)
            raw_311 = pd.concat([raw_311, raw_311_new], ignore_index=True)
            print(i," file size: ", raw_311.shape)
            display(raw_311.tail(1))
            raw_311.to_csv('s3://311-calls-etl-export/raw-noise/raw_311_noise.csv')
            break
        except:
            c+=1
        
    
```

```python
raw_311 = pd.read_csv('s3://311-calls-etl-export/raw-noise/raw_311_noise.csv')
raw_311.drop('Unnamed: 0', axis=1, inplace = True)
display(raw_311.tail(1))
raw_311.shape
```

```python
t_311 = raw_311.copy()
t_311.dropna(subset=['incident_zip'], inplace = True)
t_311['incident_zip'] = t_311['incident_zip'].astype('int')
t_311['year'] = pd.to_datetime(t_311['created_date']).dt.year
t_311['hour'] = pd.to_datetime(t_311['created_date']).dt.hour
t_311['hour_range'] = np.where(t_311['hour']<6, '0-6 hours',
                       np.where(t_311['hour']<12, '7-12 hours',
                       np.where(t_311['hour']<18, '13-18 hours', '19-24 hours')))
t_311['month'] = pd.to_datetime(t_311['created_date']).dt.year.astype('str') +'-'+pd.to_datetime(t_311['created_date']).dt.month.astype('str').str.zfill(2)
display(t_311[['year','unique_key']].groupby(by='year').count())
display(t_311[['hour_range','unique_key']].groupby(by='hour_range').count())
```

```python
t_311[['incident_zip','unique_key']].groupby(by='incident_zip').count().to_csv('zip_codes.csv')
```

```python
t_311 = t_311.join(pd.get_dummies(t_311['descriptor']))
t_311 = t_311.join(pd.get_dummies(t_311['hour_range']))
t_311.head().to_csv('head.csv')
```
