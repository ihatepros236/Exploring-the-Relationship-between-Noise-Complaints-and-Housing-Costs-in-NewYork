```python
import pandas as pd
import numpy as np
from datetime import datetime
```

# Zillow home value index

```python
zillow_index_sales = pd.read_csv('Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv')
zillow_index_sales.head()
```

```python
zip_codes_with_complaints = pd.read_csv('zip_codes.csv')
zip_codes_with_complaints.head()
```

```python
columns_zillow = list(zillow_index_sales.columns)
list_zips_311 = list(zip_codes_with_complaints['incident_zip'])
filter_zillow_sales = zillow_index_sales[zillow_index_sales['RegionName'].isin(list_zips_311)]
zillow_melt_sales = pd.melt(filter_zillow_sales, id_vars='RegionName', value_vars=[x for x in columns_zillow if ('201' in x or '202' in x) ])\
            .sort_values(by=['RegionName','variable'])\
            .rename(columns = {'variable':'month'})
zillow_melt_sales['zillow_month'] =  pd.DatetimeIndex(zillow_melt_sales['month']).year.astype('str') + '-' +pd.DatetimeIndex(zillow_melt_sales['month']).month.astype('str').str.zfill(2)
#zillow_melt_sales.drop(columns=('month'), inplace = True)
zillow_melt_sales.rename(columns={'value':'zillow_sales_value_index','month': 'zillow_month_day','RegionName':'zillow_zip'},inplace = True)
zillow_melt_sales.head()
```

```python
zillow_melt_sales.to_csv('zillow_sales_value_index_zip_month.csv')
```
